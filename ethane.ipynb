```python
from chemview import  MolecularViewer
from playground import Playground
```

```python
pg = Playground()
molecule = pg.open("h2o.cml")
```

```python
mv = MolecularViewer(molecule.coordinates, topology={'atom_types': molecule.atom_types, 
                                                     'bonds': molecule.bonds})

for orbital in molecule.orbitals:
    mv.add_isosurface(orbital, isolevel=0.05, color=0xff0000)
    mv.add_isosurface(orbital, isolevel=-0.05, color=0x0000ff)
# mv.add_isosurface(molecule, isolevel=5, color=0xff0000)

mv.ball_and_sticks()
mv
```
